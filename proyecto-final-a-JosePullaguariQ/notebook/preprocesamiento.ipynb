```python
import pandas as pd 
import matplotlib.pyplot as plt
```


### Selección de fuente de datos
#### Se selecciona la fuente de datos de vacunas diarias por provincias del Ecuador del año 2021


```python
# Lectura de datos

datos = pd.read_csv('data/provincias_por_dosis_total_diaria_acumuladas.csv',index_col="provincia") 
```

```python
datos.head()
```

```python
# Para la limpieza de datos se elimina las columnas de latitud y longitud, de las cuales no se tomará en cuenta para graficar

datos.drop(['lat', 'lng'], axis=1)
```

```python
# Gráfica que presenta vacunas de una provincia específica
provincia_Loja = datos.iloc[11,3:]
plt.figure(figsize=(20,8))
ax = plt.axes()
provincia_Loja.plot(c="#51DF12")
plt.title('Vacunas de Loja del año 2021')
ax.set_xlabel("\nFecha", size=20, color="red" )
ax.set_ylabel("Vacunas", size=20, color="red")
plt.show()
```

```python
# Gráfica que presenta el número de vacunas de provincias y una fecha específica
data1=[datos.index.values[11]]
data2=[datos.index.values[15]]

year1=[datos['01/05/2021'].values[11]]
year2=[datos['01/05/2021'].values[15]]

fig = plt.figure(figsize=(20,5))
plt.rcParams["font.size"] = 18

ax0 = fig.add_subplot(1,2,1) 
ax1 = fig.add_subplot(1,2,2)

ax0.bar(data1, year1, color="red")
ax0.legend([datos['01/05/2021'].values[11]])
ax0.set_xlabel("\nProvincia", size=20, color="#7E0E24")
ax0.set_ylabel("Vacunas", size=20, color="#7E0E24")

ax1.bar(data2, year2, color="blue")
ax1.legend([datos['01/05/2021'].values[15]])
ax1.set_xlabel("\nProvincia", size=20, color="#7E0E24")
ax1.set_ylabel("Vacunas", size=20, color="#7E0E24")
plt.show()
```

```python
# Gráfica que presenta comparación de vacunas entre provincias en una fecha específica
eje_x =  [datos.index.values[11], datos.index.values[15]]
eje_y = [datos['01/09/2021'].values[11], datos['01/09/2021'].values[15]]


plt.figure(figsize=(12,5))
plt.bar(eje_x, eje_y, width=0.5, color="#CF0C32" )

plt.ylabel('Vacunas al día')
plt.xlabel('Provincia')

ax = plt.axes()
plt.title('Vacunas del 1 de septiembre del 2021')
ax.set_xlabel("\nProvincia", size=20, color="#7E0E24")
ax.set_ylabel("Vacunas", size=20, color="#7E0E24")

plt.show()
```

```python
# Gráfica que presenta vacunas de dos fechas específicas
datos.loc[:,['01/02/2021','01/03/2021']].plot(figsize=(20, 8))
ax = plt.axes()
plt.title('Vacunas de los dias 1 de febrero y 1 de marzo del 2021')
ax.set_xlabel("\nProvincia", size=20, color="#1D888D")
ax.set_ylabel("Vacunas", size=20, color="#1D888D")
```

```python
# Gráfica que presenta las provincias con su población
df_reset = datos.reset_index()
fig, ax = plt.subplots(figsize=(20,8))
plt.barh(df_reset.provincia, df_reset.poblacion, color="#C7894F" )
plt.xlabel("Población", size=20, color="#A25813")
plt.ylabel("Provincia", size=20, color="#A25813")
plt.show()
```
